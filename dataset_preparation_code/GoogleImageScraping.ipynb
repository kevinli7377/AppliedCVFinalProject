```python
from PIL import Image, ImageOps
import os
import math

image_directory = '/content/drive/MyDrive/Applied CV Project/google-image-scraper/images/pasta'

output_directory = '/content/drive/MyDrive/Applied CV Project/google-image-scraper/images/pasta'

os.makedirs(output_directory, exist_ok=True)

rotation_degrees = 2
shear_degrees = 1
shear_degrees_2 = 2
shear_degrees_3 = 3

def save_image(img, original_filename, operation_name):
    name, ext = os.path.splitext(original_filename)
    new_filename = f"{name}_{operation_name}{ext}"
    img.save(os.path.join(output_directory, new_filename))
    print(f"Processed and saved: {new_filename}")


# Iterate through for 1) Flipping, 2) Rotating, 3) Shearing
for filename in os.listdir(image_directory):
    if filename.endswith(('.png', '.jpg', '.jpeg')):
        original_path = os.path.join(image_directory, filename)
        print(original_path)

        with Image.open(original_path) as img:

            # 1. Flipping: Flip the image along Y-axis and save
            # flipped_img = ImageOps.mirror(img)
            # save_image(flipped_img, filename, "flipped")

            # 2. Rotate Rotate the image slightly and save
            rotated_img = img.rotate(rotation_degrees, expand=True, fillcolor='white')
            save_image(rotated_img, filename, "rotated")

            # 3. Shear the image slightly and save
            # Affine transform
            shear = math.tan(math.radians(shear_degrees))
            a = 1
            b = 0
            c = 0  # Left/right translation
            d = shear  # Shear
            e = 1
            f = 0  # Up/down translation
            sheared_img = img.transform(img.size, Image.AFFINE, (a, b, c, d, e, f), fillcolor='white')
            save_image(sheared_img, filename, "sheared")

            d = math.tan(math.radians(shear_degrees_2))

            sheared_img = img.transform(img.size, Image.AFFINE, (a, b, c, d, e, f), fillcolor='white')
            save_image(sheared_img, filename, "sheared2")

            d = math.tan(math.radians(shear_degrees_3))

            sheared_img = img.transform(img.size, Image.AFFINE, (a, b, c, d, e, f), fillcolor='white')
            save_image(sheared_img, filename, "sheared3")

    else:
        print(f"Skipping file: {filename}")
```

```python
# Iterate through all the images and rename them
from PIL import Image
import os

image_directory = '/content/drive/MyDrive/Applied CV Project/google-image-scraper/images/pasta'

file_counter = 1

for filename in sorted(os.listdir(image_directory)):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
        original_path = os.path.join(image_directory, filename)

        new_filename = f"{file_counter}.jpg"
        new_path = os.path.join(image_directory, new_filename)

        try:
            with Image.open(original_path) as img:
                rgb_im = img.convert('RGB')
                rgb_im.save(new_path, 'JPEG')
                os.remove(original_path)  # This line deletes the original file
            print(f"Renamed '{filename}' to '{new_filename}'")
        except Exception as e:
            print(f"Error processing '{filename}': {e}")

        file_counter += 1
```
