```python
import os
import shutil
from sklearn.model_selection import train_test_split

# For freiburg groceries scene

data_dir = '/content/drive/MyDrive/Applied CV Project/freiburg_groceries_scenes/images'
file_names = [f for f in os.listdir(data_dir) if os.path.isfile(os.path.join(data_dir,f))]

train_files, test_files = train_test_split(file_names, test_size = 0.2, random_state=42)

train_dir = os.path.join(data_dir,'train')
test_dir = os.path.join(data_dir,'test')
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

for file in train_files:
    shutil.copy(os.path.join(data_dir, file), train_dir)

for file in test_files:
    shutil.copy(os.path.join(data_dir, file), test_dir)

print("Data split complete. Check the 'train' and 'test' folders.")
```

```python
train_txt_path = os.path.join(data_dir, 'train_files.txt')
test_txt_path = os.path.join(data_dir, 'test_files.txt')

with open(train_txt_path, 'w') as train_txt:
    for file in train_files:
        train_txt.write("%s\n" % file)

with open(test_txt_path, 'w') as test_txt:
    for file in test_files:
        test_txt.write("%s\n" % file)

print("Text files listing the train and test images have been created")
```

For freiburg_groceries_dataset

```python
import os
import shutil
from sklearn.model_selection import train_test_split

# For freiburg groceries scene

data_dir = '/content/drive/MyDrive/Applied CV Project/freiburg_groceries_dataset'

def create_train_test(data_dir):
  print(data_dir)
  file_names = [f for f in os.listdir(data_dir) if os.path.isfile(os.path.join(data_dir,f))]

  train_files, test_files = train_test_split(file_names, test_size = 0.2, random_state=42)

  train_dir = os.path.join(data_dir,'train')
  test_dir = os.path.join(data_dir,'test')
  os.makedirs(train_dir, exist_ok=True)
  os.makedirs(test_dir, exist_ok=True)

  for file in train_files:
      shutil.copy(os.path.join(data_dir, file), train_dir)

  for file in test_files:
      shutil.copy(os.path.join(data_dir, file), test_dir)

  print("Data split complete. Check the 'train' and 'test' folders.")
```

```python
def create_txt(data_dir):
  train_path = os.path.join(data_dir, 'train/')
  test_path = os.path.join(data_dir, 'test/')

  train_txt_dir = os.path.join(data_dir, 'train/txt/')
  test_txt_dir = os.path.join(data_dir, 'test/txt/')

  train_txt_path = os.path.join(data_dir, 'train/txt/train.txt')
  test_txt_path = os.path.join(data_dir, 'test/txt/test.txt')

  os.makedirs(train_txt_dir, exist_ok=True)
  os.makedirs(test_txt_dir, exist_ok=True)

  train_files = [f for f in os.listdir(train_path) if os.path.isfile(os.path.join(train_path, f))]
  test_files = [f for f in os.listdir(test_path) if os.path.isfile(os.path.join(test_path, f))]


  with open(train_txt_path, 'w') as train_txt:
    for file in train_files:
        train_txt.write("%s\n" % file)

  with open(test_txt_path, 'w') as test_txt:
      for file in test_files:
          test_txt.write("%s\n" % file)
```

```python
category_folders = [f for f in os.listdir(data_dir) if os.path.isdir(os.path.join(data_dir, f))]

for category in category_folders:
  category_rel_path = os.path.join(data_dir, category)
  create_train_test(category_rel_path)
  create_txt(category_rel_path)
```

```python
import json
from sklearn.model_selection import train_test_split

json_path = '/content/drive/MyDrive/Applied CV Project/d2s_annotations_extracted/d2s_annotations_v1.1/processed_annotations/D2S_super_processed.json'

with open(json_path, 'r') as file:
  original_data = json.load(file)

modified_data_json = []

for image_id, labels in original_data.items():
  img_file_path = f"D2S_{int(image_id):06}.jpg"
  file_path = f"./images/{img_file_path}" # Need to cd into ds2_images_extracted

  modified_data_json.append({
      "image_id":image_id,
      "file_path":file_path,
      "labels": list([label.replace("_", " ").lower() for label in labels])
  })

train_items, test_items = train_test_split(modified_data_json, test_size=0.2, random_state=42)

with open('/content/drive/MyDrive/Applied CV Project/d2s_images_extracted/train.json','w') as file:
  json.dump(train_items, file)

with open('/content/drive/MyDrive/Applied CV Project/d2s_images_extracted/test.json', 'w') as file:
  json.dump(test_items, file)
```

```python
# Create second version where labels are sets
# i.e., if a scene has multiple of the same item,
# the labels will only have one instance of that item as opposed to many
# If a scene has 3 oranges, labels will be ['Orange'] as opposed to ['Orange','Orange','Organce']

import json
from sklearn.model_selection import train_test_split

json_path = '/content/drive/MyDrive/Applied CV Project/d2s_annotations_extracted/d2s_annotations_v1.1/processed_annotations/D2S_super_processed.json'

with open(json_path, 'r') as file:
  original_data = json.load(file)

modified_data_json = []

for image_id, labels in original_data.items():
  img_file_path = f"D2S_{int(image_id):06}.jpg"
  file_path = f"./images/{img_file_path}" # Need to cd into ds2_images_extracted

  modified_data_json.append({
      "image_id":image_id,
      "file_path":file_path,
      "labels": list(set([label.replace("_", " ").lower() for label in labels]))
  })

train_items, test_items = train_test_split(modified_data_json, test_size=0.2, random_state=42)

with open('/content/drive/MyDrive/Applied CV Project/d2s_images_extracted/train_uniqueLabels.json','w') as file:
  json.dump(train_items, file)

with open('/content/drive/MyDrive/Applied CV Project/d2s_images_extracted/test_uniqueLabels.json', 'w') as file:
  json.dump(test_items, file)
```

```python
import os
import shutil
from sklearn.model_selection import train_test_split

root_dir = '/content/drive/MyDrive/Applied CV Project/fruits-360_dataset/images'


directories = ['beef']

print(directories)
# Iterate through all directories and subdirectories in the root_dir
for dir_name in directories:

      dir_path = os.path.join(root_dir, dir_name)
      file_names = [f for f in os.listdir(dir_path) if os.path.isfile(os.path.join(dir_path, f))]
      print(file_names)
      print(dir_path)

      if not file_names:
          continue

      train_files, test_files = train_test_split(file_names, test_size=0.2, random_state=42)

      # train and test directories
      train_dir = os.path.join(dir_path, 'train')
      test_dir = os.path.join(dir_path, 'test')

      # Create train and test directories if they don't exist
      os.makedirs(train_dir, exist_ok=True)
      os.makedirs(test_dir, exist_ok=True)

      # Copy the images to the train and test directories
      for file in train_files:
        shutil.copy(os.path.join(dir_path, file), train_dir)

      for file in test_files:
        shutil.copy(os.path.join(dir_path, file), test_dir)

      # break

print("Data split complete. Check the 'train' and 'test' folders in each category.")
```
